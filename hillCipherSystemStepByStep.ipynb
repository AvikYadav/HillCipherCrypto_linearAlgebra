```python
#hill cipher forward pass
import numpy as np
```

```python
mapping_dict = {
    "a": 0,
    "b": 1,
    "c": 2,
    "d": 3,
    "e": 4,
    "f": 5,
    "g": 6,
    "h": 7,
    "i": 8,
    "j": 9,
    "k": 10,
    "l": 11,
    "m": 12,
    "n": 13,
    "o": 14,
    "p": 15,
    "q": 16,
    "r": 17,
    "s": 18,
    "t": 19,
    "u": 20,
    "v": 21,
    "w": 22,
    "x": 23,
    "y": 24,
    "z": 25
}
reverse_mapping = {v:k for k,v in mapping_dict.items() }
```

```python
raw_text = "Hello this is hill cipher encryption project"

stripped_space_text = raw_text.lower().replace(" ","")
```

```python
#breaking text into pieces of 2
if len(stripped_space_text)%2 != 0:stripped_space_text+="x"

splitted_text_into_chunks = np.array([list(stripped_space_text[i-2:i]) for i in range(2,len(stripped_space_text)+1,2)])
mapping_chunks_to_numbers = np.array([list(map(lambda x:mapping_dict[str(x)],i)) for i in splitted_text_into_chunks]).T
print(mapping_chunks_to_numbers)
```

```python
key = np.array([[3,3],[2,5]])

applying_key_to_each_vector = []

for index in range(len(mapping_chunks_to_numbers[0])):
    column = mapping_chunks_to_numbers[:,index].reshape(2,-1)
    applying_key = (key@column)%26
    applying_key_to_each_vector.append(applying_key)
```

```python
converted_back_to_letters = []
for vector in applying_key_to_each_vector:
    converted = []
    for value in vector:
        converted.append(reverse_mapping[int(value[0])])
    converted_back_to_letters.append(converted)

print(converted_back_to_letters)
```

```python
string_encrypted = ""
for each in converted_back_to_letters:
    string_encrypted += "".join(each)

print(string_encrypted)
```

```python
def gcd(numa,numb):
    numa,numb = int(numa),int(numb)
    factorsa = [i for i in range(1,numa) if numa%i ==0]
    factorsb = [i for i in range(1,numb) if numb%i ==0]
    intersection = set(factorsa).intersection(set(factorsb))
    return intersection == {1}
```

```python
#finding key inverse matrix
# 9x = 26k + 1
def modulous_inverse(n):
    k=0
    while True:
        if (26*k+1)%n==0:
            return (26*k+1)/n
        k+=1

def find_modulus_inverse_of_matrix(matrix):
    ad_matrix = (np.linalg.inv(matrix)*np.linalg.det(matrix))%26
    det_key = round(np.linalg.det(matrix),2)
    if not gcd(det_key,26):
        return "invalid"
    modulus_inverse_det_key = modulous_inverse(det_key)
    mat_inverse = (modulus_inverse_det_key*ad_matrix)%26
    return mat_inverse

key_inverse = find_modulus_inverse_of_matrix(key)
verify = (key@key_inverse)%26
print(verify)
```
